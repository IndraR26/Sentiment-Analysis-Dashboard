```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
#Load Dataset
file_path = "Sentiment-Analysis-Dashboard\Dataset\PRDECT-ID Dataset.csv"
df = pd.read_csv(file_path)
```

```python
# Kata kunci positif dan negatif
positive_keywords = ["baik", "ramah", "cepat", "amanah", "terima kasih", "bagus", "mantap",
                         "paten", "joss", "pas", "puas", "senang", "aman", "memuaskan", "terbaik", "sesuai", "rapi", "berkualitas", "love", "alhamdulillah"]
negative_keywords = ["buruk", "rusak", "jelek", "lambat", "tidak", "kurang", "parah", "mengecewakan",
                         "lama", "salah", "cacat", "buruk", "penyok", "masalah", "ketipu", "error", "boros", "kw", "palsu", "bosok", "lecet", "hancur", "bad", "payah", "penipuan", "terlambat"]

# Ekstraksi kata dengan koma sebagai delimiter
def extract_keywords(text, keywords):
    words = re.findall(r'\w+', text.lower())
    return ','.join([word for word in words if word in keywords])

# Tambahkan kolom positif dan negatif yang dipisahkan dengan koma
df['Positive Words'] = df['Customer Review'].apply(lambda x: extract_keywords(str(x), positive_keywords))
df['Negative Words'] = df['Customer Review'].apply(lambda x: extract_keywords(str(x), negative_keywords))

# Simpan ke CSV
df.to_csv("processed_reviews.csv", index=False)
print("Data berhasil disimpan.")
```
